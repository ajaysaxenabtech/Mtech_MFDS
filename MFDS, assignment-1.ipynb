```python
#(i) function to check whether a given square matrix is diagonally dominant.
# Deliverable(s) : Code that performs the check and the results obtained for the matrix

import numpy as np

def dom(m,n) :
 
    for i in range(0, n) :        
        sum = 0
        for j in range(0, n) :
            sum = sum + abs(m[i][j])    
 
        sum = sum - abs(m[i][i])
        if(abs(m[i][i]) > sum or abs(m[i][i]) == sum):
            True
        
        return False
 
n = 4
m =np.random.randint(-10,10,size=(n,n))
print(m,"\n")
if((dom(m,n))) :
    print ("This is a Diagonally Dominant Matrix")
else :
    print ("This is not a Diagonally Dominant Matrix") 
```
